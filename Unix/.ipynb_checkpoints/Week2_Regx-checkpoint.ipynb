# Search
## Regular Expressions

- `grep`

<pre>grep "x" states.txt</pre>

- metacharacters: characters that can represent other characters
    - `egrep` extends `grep`’s capabilities
    -  "`.`" (period) metacharacter, which represents any character
    <pre>
    ## search states.txt for the character “i”, 
    ## followed by any character, followed by the character “g”
    egrep "i.g" states.txt
    ## Virginia
    ## Washington
    ## West Virginia
    ## Wyoming</pre>
    
    >`quantifiers` allow you to specify the number of times a particular regular expression should appear in a string
    - "`+`" (plus), basic quantifiers, represents one or more occurrences of the preceeding expression.
    >  “`s+as`” means: one or more “`s`” followed by “`as`”
    
    <pre>
    egrep "s+as" states.txt
    ## Arkansas
    ## Kansas
    </pre>
    
    - "`*`" (star) metacharacter which represents zero or more occurrences of the preceding expression    
    <pre>
    egrep "s*as" states.txt
    ## Alaska
    ## Arkansas
    ## Kansas
    ## Massachusetts
    ## Nebraska
    ## Texas
    ## Washington
    </pre>
    
    - "`{ }`" curly brackets, specify an exact number of occurrences of an expression
    <pre>
    egrep "s{2}" states.txt
    ## Massachusetts
    ## Mississippi
    ## Missouri
    ## Tennessee
    </pre>
    
    > "s{2}" == "ss" (exactly two occurrences)
    
    <pre>
    egrep "s{2,3}" states.txt
    </pre>
    
    > "`s{2,3}`": state names that have between two and three adjacent occurrences of the letter “s” 
    
    - "`( )`": create capturing groups within regular expressions. You can use a capturing group in order to search for multiple occurrences of a string. 
    
    <pre>
    egrep "(iss){2}" states.txt
    ## Mississippi</pre>
    
    > search for the string “iss” occurring twice in a state name
    
    - We could combine more quantifiers and capturing groups to dream up even more complicated regular expressions
    <pre>
    egrep "(i.{2}){3}" states.txt
    ## Mississippi
    </pre>
    
    > three occurrences of an “i” followed by two of any character
